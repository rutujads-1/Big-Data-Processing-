```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
df = pd.read_csv("/content/trip_cnts.csv")

# Calculate 'business-month'
df['business-month'] = df['business'] + '-' + df['month'].astype(str)

# Plot histogram with logarithmic scale on y-axis
plt.figure(figsize=(10, 6))
bars = plt.bar(df['business-month'], df['trip_count'])
plt.yscale('log')  # Set y-axis scale to logarithmic
plt.xlabel('Business-Month')
plt.ylabel('Trip Counts (log scale)')
plt.title('Histogram of Trip Counts by Business-Month')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Add text labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), va='bottom', ha='center')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read data from CSV file
df = pd.read_csv("/content/profits_month.csv")

# Calculate 'business-month' column
df['business-month'] = df['business'] + '-' + df['month'].astype(str)

# Plotting
plt.figure(figsize=(10, 6))
colors = ['blue', 'green', 'red', 'cyan', 'magenta']  # Define colors for each business
bars = plt.bar(df['business-month'], df['total_profit'], color=colors)
plt.xlabel('Business-Month')
plt.ylabel('Total Profit')
plt.title('Total Profit for Each Business in Each Month')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Add text labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{height:.2f}', ha='center', va='bottom')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
df = pd.read_csv('/content/earning_month.csv')

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar([f"{row['business']}-{row['month']}" for _, row in df.iterrows()], df['total_earnings'], color='skyblue')
plt.xlabel('Business-Month')
plt.ylabel("Driver's Earnings")
plt.title("Driver's Earnings for Each Business in Each Month")
plt.xticks(rotation=45, ha='right')
plt.yscale('log')  # Set y-axis scale to logarithmic
plt.tight_layout()

# Add text labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height, '%.2f' % height, ha='center', va='bottom')

plt.show()
```
